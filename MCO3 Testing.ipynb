## These testing module contains perversed/contrived, and random test data. Complete test cases for the IMDB movies dataset is included in MCO5 documentation.

### The IMDB movie dataset is converted into 2 dimensions using PCA. It was chosen since other dimension reduction techniques were slow

### Only test cases with 2 dimensions are plotted in a scatter plot. Other multidimensional are just clustered using different setups of kmeans


### In each test cases, frequency counts of differnet k-means setups (Elkan + k-means++, Lloyd + k-means++, Lloyd + random) are printed. Next a plot that shows the clustered data using Elkan's k-means will be shown.

### For Elkan's kmeans, the code used is the modified sklearn's k-means located in the main MCO3 document
### For the others, the original sklearn's k-means implementations were used. The number of comparison were just computed using the formula mentioned in main MCO3 document

### Same Values

```python

X = np.zeros((40000, 2))
X[0:10000, :] = 30.0
X[10000:20000, :] = 60.0
X[20000:30000, :] = 90.0
X[30000:, :] = 120.0
n_clusters = 4

clusters = doTests(n_clusters, X)
df = pd.DataFrame(X)
df['cluster'] = clusters

plot(df)
```

### Uniform permutation of numbers but few

```python
size = 10
X = np.zeros((size * size, 2))
counter = 0
for i in range(size):
    for j in range(size):
        X[counter][0] = i
        X[counter][1] = j
        counter += 1

clusters = doTests(n_clusters, X)
df = pd.DataFrame(X)
df['cluster'] = clusters

plot(df)
```

### Uniform 3D data

```python
size = 30
n_clusters = 9

X = np.zeros((size * size * size, 3))
counter = 0
for i in range(size):
    for j in range(size):
        for k in range(size):
            X[counter][0] = i
            X[counter][1] = j
            X[counter][2] = k
            counter += 1

clusters = doTests(n_clusters, X)
df = pd.DataFrame(X)
df['cluster'] = clusters
print("Number of members in each cluster")
df['cluster'].value_counts()
```
